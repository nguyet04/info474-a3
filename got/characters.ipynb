```python
locations_numScenes = {}
```

```python
import datetime
def time_sum(time1, time2):
  a = str_to_timedelta(time1)
  b = str_to_timedelta(time2)
  return str(a + b)

def time_difference(start, end):
  start_delta = str_to_timedelta(start)
  end_delta = str_to_timedelta(end)
  return str(end_delta - start_delta)

def str_to_timedelta(string):
  (h, m ,s) = string.split(':')
  d = datetime.timedelta(hours = int(h), minutes=int(m), seconds=int(s))
  return d
```

```python
# take in the season and create 
def createJson(inputSeason):
    locations_numScenes = {}
    locations_numDeaths = {}
    locations_screenTime = {}
    for episode in data:
        if episode['seasonNum'] == inputSeason:
            for scene in episode['scenes']:
                duration = time_difference(scene['sceneStart'], scene['sceneEnd'])
                currLocation = scene['location']
                if currLocation not in locations_numScenes:
                    locations_numScenes[currLocation] = 1
                else:
                    locations_numScenes[currLocation] += 1
                
                if currLocation in locations_screenTime:
                    locations_screenTime[currLocation] = time_sum(locations_screenTime[currLocation], duration)
                else:
                    locations_screenTime[currLocation] = duration
              
                for character in scene['characters']:
                    if 'alive' in character:
                        if character['alive'] == False:
                            currLocation = scene['location']
                            if currLocation not in locations_numDeaths:
                                locations_numDeaths[currLocation] = 1
                            else: 
                                locations_numDeaths[currLocation] += 1
    outputJson = []
    for location in locations_numScenes:
        numDeaths = 0
        if location in locations_numDeaths:
            numDeaths = locations_numDeaths[location]
        outputJson.append(
            {
                "location": location,
                "season": inputSeason,
                "numOfScenes": locations_numScenes[location],
                "numOfDeaths": numDeaths,
                "screenTime": locations_screenTime[location]
            }
        )
                            
    return outputJson
        
```

```python
createJson(1)
```

```python
final_output = []

for i in range(8):
    final_output.append(createJson(i + 1))
```

```python
final_output
```

```python
with open('locations.json', 'w') as outfile:
    json.dump(final_output, outfile)
```
